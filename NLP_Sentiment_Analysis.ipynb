```python
import numpy as np

import pandas as pd
```

```python
# Read CSV file given as an Input from the user. Here, I took index column as ID

df = pd.read_csv("chrome_reviews.csv",index_col='ID')

df
```

#### - We can see that all the columns are non-null except Text,Developer Reply and Version.
     
#### - But we can see Some columns are not relevant to our analysis.So, we will drop those columns.

```python
df.drop(['User Name','Developer Reply','Thumbs Up','Version','Review Date','App ID','Review URL'] ,axis = 1 ,inplace = True)
```

```python
df
```

### Check for any blank row and if it is there then we will drop it.

```python
df.dropna(inplace=True)

blanks = []  # start with an empty list

for i,lb,rv in df.itertuples():  # iterate over the DataFrame
    if type(rv)==str:            # avoid NaN values
        if rv.isspace():         # test 'review' for whitespace
            blanks.append(i)     # add matching index numbers to the list

df.drop(blanks, inplace=True)
```

### Final filtered data for processing

```python
df
```

### Importing libraries for sentiment analysis to understand the review.

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

### Assigning scores for each review

```python
df['scores'] = df['Text'].apply(lambda review: sid.polarity_scores(review))

df['compound']  = df['scores'].apply(lambda score_dict: score_dict['compound'])

df['comp_score'] = df['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')

df
```

### Filtering out worst rated movie on the basis of its compound score

### Note: Here, our aim is to take out those movies where comp_score is positive but still star is 1.



```python
df_worst_rated = df[(df.loc[:,'Star'] == 1) & (df.loc[:,'comp_score'] == 'pos') & (df.loc[:,'compound'] >0)]
```

### Filtered movies where review does not justify the ratings

```python
df_worst_rated
```

### So, there are total 382 such users ID who gave incorrect rating

```python
df_worst_rated.shape
```

```python
df_incorrect_reviews = df_worst_rated.drop(['scores','comp_score','compound'],axis= 1)
```
